```python
b_ini = 0.0
w_ini = np.array([0.0, 0.0, 0.0, 0.0])
```

```python
#cost function
m=x_train.shape[0]
for i in range(m):
    f_wb=np.dot(w_ini,x_train[i])+b_ini
    cost=(np.sum((f_wb-y_train[i])**2))
cost_f=(1/(2*m))*cost
```

```python
def gradient(x,y,w,b):
    m,n=x.shape
    dj_dw = np.zeros((n,))
    dj_db = 0
    for i in range(m):
        err=((np.dot(x[i],w)+b)-y[i])
        dj_db=dj_db+err
        for j in range(n):
            dj_dw[j]=dj_dw[j]+err*x[i,j]
    dj_dw=dj_dw/m
    dj_db=dj_db/m
    return dj_db,dj_dw
```

```python
db,dw=gradient(x_train,y_train,w_ini,b_ini)
```

```python
p=0.01
for i in range(123):
    w_ini=w_ini-p*dw
    b_ini=b_ini-p*db
    for j in range(m):
        f_wb=np.dot(w_ini,x_train[j])+b_ini
        cost=(np.sum((f_wb-y_train[j])**2))
    cost_f=(1/(2*m))*cost
    print(f"w:{w_ini} b:{b_ini} cost:{cost_f} iteration:{i}")
print(f"the final value is w:{w_ini} b:{b_ini} cost:{cost_f}")
```
